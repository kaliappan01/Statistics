```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
no_of_trials = [2, 5, 20 ,50 ,100 ,1000]
data = stats.bernoulli.rvs(0.5, size = no_of_trials[-1])
# print(data)
x = np.linspace(0,1,100)
for i, N in enumerate(no_of_trials):
  tails = data[:N].sum()
  ax = plt.subplot(len(no_of_trials)/2, 2, i+1)
  ax.set_title("{} trials, {} Heads".format(N, tails))
  plt.xlabel("Probability of Heads")
  plt.ylabel("Density")
  if i==0:
    plt.ylim([0.0, 2.0])
  plt.setp(ax.get_yticklabels(), visible = False)
  y = stats.beta.pdf(x, 1+tails, 1+N-tails)
  plt.plot(x, y, label = "Observed {} tosses, \n {} heads".format(N, tails))
  plt.fill_between(x, 0, y, color="#29c685", alpha = 0.5)
plt.tight_layout()
```
